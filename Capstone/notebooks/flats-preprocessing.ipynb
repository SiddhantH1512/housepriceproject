```python
# DROPPING UNNECESSARY COLUMNS
print(f"Shape of dataframe before dropping columns {df.shape}")

df.drop(columns=["link", "description", "property_id"], inplace=True)

print(f"Shape of dataframe after dropping columns {df.shape}")
```

```python
# CONVERTING PRICE COLUMN INTO ONE SINGLE UNIT (CRORES)
df["price"].value_counts()
```

We are not concerned with price on request value as it wont be helpful for our regression model. So we need to remove that entry from the price column.

```python
df = df[df["price"] != "Price on Request"]

def price(value):
    if type(value) == float:
        return value
    else:
        if value[1] == "Lac":
            return round(float(value[0])/100, 2)
        else:
            return round(float(value[0]),2)
   
    

df["price"] =  df["price"].str.split(" ").apply(price)
df.sample(5)
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df["society"].value_counts()
```

Majority of the society names have a rating and a star attached to them. We need to clean this column by removing this rating and converting all names to lower case to bring the in a standard format.

```python
# CLEANING UP THE 'SOCIETY' COLUMN
def regex(value):
    cleaned_value = re.sub(r'\d+(\.\d+)?\s?★', "", str(value))
    cleaned_value = cleaned_value.lower().strip()
    return cleaned_value

df["society"] = df["society"] .apply(regex)
df.sample(5)
```

```python
# CLEANING UP AREA COLUMN - RENAMING THE COLUMN
df.rename(columns={"area":"price_per_sqft"}, inplace=True)
```

```python
# CLEANING UP AREA COLUMN - JUST KEEPING THE VALUE
df["price_per_sqft"] = df["price_per_sqft"].str.split("/").str.get(0).str.replace("₹", "").str.replace(",","").str.strip().astype(float)
df.sample(5)
```

```python
# NOW SINCE WE HAVE price_per_sqft AND price WE CAN ACTUALLY CREATE A FETAURE CALLED AREA.
df.insert(loc=4, column="area", value=round((df["price"] * 10000000)/df["price_per_sqft"],2))
df.insert(loc=2, column="property_type", value="flat")

df.head(5)
```

```python
# CLEANING THE BEDROOM COLUMN
print(f"{df['bedRoom'].value_counts()}")
print(f"Number of missing values in bedroom column are: {df['bedRoom'].isnull().sum()}")
```

No Missing values in the bedroom column. We can go ahead and clean it

```python
df["bedRoom"] = df["bedRoom"].str.split(" ").str.get(0).str.strip().astype(int)
df.sample(3)
```

```python
# CLEANING THE BATHROOM COLUMN
print(f"{df['bathroom'].value_counts()}")
print(f"Number of missing values in bathroom column are: {df['bathroom'].isnull().sum()}")
```

No Missing values in the bathroom column. We can go ahead and clean it

```python
df["bathroom"] = df["bathroom"].str.split(" ").str.get(0).str.strip().astype(int)
df.sample(3)
```

```python
# CLEANING THE BALCONY COLUMN
print(f"{df['balcony'].value_counts()}")
print(f"Number of missing values in balcony column are: {df['balcony'].isnull().sum()}")
```

No Missing values in the bathroom column. We can go ahead and clean it. We are not converting this column as interger because we have a category as 3+

```python
df["balcony"] = df["balcony"].str.split(" ").str.get(0).str.strip()
df.sample(3)
```

```python
# LETS LOOK INTO THE ADDITIONALROOM COLUMN
print(f"{df['additionalRoom'].value_counts()}")
print(f"Number of missing values in additionalRoom column are: {df['additionalRoom'].isnull().sum()}")
```

It looks like there are multiple categories in this column and also there are 1305 null values. So as of now lets just replace these null values with "not available" and convert everything to lower case. We will deal with the categories later.

```python
df["additionalRoom"].fillna("not available", inplace=True)
df["additionalRoom"] = df["additionalRoom"].str.lower()
df.head(3)
```

```python
# LETS CLEAN THE FLOOR NUM COLUMN. WE WILL EXTRACT THE FLOOR NUMBER. (4TH OF 4 FLOORS - WE ARE INTERESTED IN 4)
df["floorNum"].value_counts()
```

There are few columns which are not actual floor numbers. They are Ground and Basement. We will convert them into 0 and -1 respectively.

```python
df["floorNum"] = df["floorNum"].str.split(" ").str.get(0).replace("Ground", "0").str.replace("Basement", "-1").str.replace('Lower', "0").str.extract(r'(\d+)nd')
df.sample(5)
```

```python
# IN THE FACING COLUMN LETS CHECK IF THERE ARE ANY NULL VALUES IF YES THEN LETS JUST REPLACE THEM WITH NA AS OF NOW
print(f"{df['facing'].value_counts()}")
print(f"{df['facing'].isnull().sum()}")
```

```python
df["facing"].fillna("NA", inplace=True)
```
